```python
def term_analysis_region(df, region=''):
    """Return the dataframe specified by each reasons for long and short term user"""
    
    reasons = pd.DataFrame({'razon': input_df['razon'].unique(), 'long':0, 'short':0})
    if region:
        df = df[df['region']==region]
        
    term_df = pd.pivot_table(df, values='uid', index=['razon'], columns=['term'], aggfunc='count').reset_index()
    term_df = pd.concat([term_df, reasons]).groupby('razon').sum().reset_index()
    term_df = term_df[term_df['razon'] != 'sin motivo']
    term_df['%_long'] = term_df['long']/term_df['long'].sum()* 100
    term_df['%_short'] = term_df['short']/term_df['short'].sum() * 100
    term_df.fillna(0, inplace=True)
    return term_df[['razon','long', '%_long', 'short', '%_short']]
    
```

```python
def term_analysis_region_detailed(df, region='', term='long'):
    """Return the dataframe specified by each reasons for long and short term user"""
    reasons = pd.DataFrame({'razon': input_df['razon'].unique(), 'Comercio':0, 'Institución':0, 'Restaurante':0})
    
    df = df[df['segmento'] != 'CNPJ Invalido']
    if region:
        df = df[df['region']==region]
        
    term_df = pd.pivot_table(df[df['term']==term], values='uid', index=['razon'], columns=['segmento'], aggfunc='count').reset_index()
    term_df = pd.concat([term_df, reasons]).groupby('razon').sum().reset_index()
    term_df = term_df[term_df['razon'] != 'sin motivo']
    term_df.fillna(0, inplace=True)
    term_df.rename(columns={
        'Comercio': 'Comercio_{}'.format(term),  
        'Institución': 'Institución_{}'.format(term), 
        'Restaurante': 'Restaurante_{}'.format(term)}, inplace=True)
    return term_df
    
```

```python
def total_analysis_region_detailed(df, region=''):
    long = term_analysis_region_detailed(df, region=region, term='long')
    short = term_analysis_region_detailed(df, region=region, term='short')
    result = pd.concat([long, short[['Comercio_short', 'Institución_short', 'Restaurante_short']]], axis=1)
    return result
```

```python
def define_quantiles(df, region=''):
    if region:
        df = df[df['region'] == region]
    
    quantiles = [
        df['subtotal'].quantile(0.25),
        df['subtotal'].quantile(0.5),
        df['subtotal'].quantile(0.75),
        df['subtotal'].quantile(1),
    ]
    
    return quantiles
```

```python
def quantiles_analysis(df, cities):
    data = {}
    data['GLOBAL'] = define_quantiles(df)
    for city in cities:
        data[city] =  define_quantiles(df, city)
    
    return pd.DataFrame(data, index=['Q1','Q2','Q3','Q4'])
```

```python
quantiles_analysis(last_month_details, cities)
```

```python
def segment_quantile_analysis(present_df, past_df, city=''):
    if city:
        present_df = present_df[present_df['region']==city]
        past_df = past_df[past_df['region']==city]
    
    present_resume = present_df.groupby(['segmento'])['uid'].count().reset_index()
    past_resume =  past_df.groupby(['segmento']).count().reset_index().rename(columns={'segmento':'segmento_past', 'uid':'uid_past'})
    
    result = pd.concat([present_resume, past_resume], axis=1)[['segmento', 'uid', 'uid_past']]
    result = result.append(result.sum(numeric_only=True), ignore_index=True)
    result['%'] = result['uid']/result['uid_past'] * 100
    result['segmento'].fillna('Total', inplace=True)
    return  result
```

```python
def quantile_assign(value, quantiles):
    if value <= quantiles[0]:
        return '0 - Q1'
    elif value > quantiles[0] and value <= quantiles[1]:
        return 'Q1 - Q2'
    elif value > quantiles[1] and value <= quantiles[2]:
        return 'Q2 - Q3'
    else:
        return 'Q3 - Q4'
```

```python
input_df['subtotal'].apply(quantile_assign, quantiles=define_quantiles(last_month_details)).value_counts()
```

```python
last_month_details['subtotal'].apply(quantile_assign, quantiles=define_quantiles(last_month_details)).value_counts()
```

```python
def quantile_segment_analysis(present_df, past_df, city=''):
    if city:
        present_df = present_df[present_df['region'] == city]
        past_df = past_df[past_df['region']== city ]
        
    actual = present_df['subtotal'].apply(quantile_assign, quantiles=define_quantiles(past_df, city)).value_counts()
    past = past_df['subtotal'].apply(quantile_assign, quantiles=define_quantiles(past_df, city)).value_counts()
    result = pd.concat([actual, past], axis=1)
    result = result.append(result.sum(numeric_only=True), ignore_index=True)
    result['%'] =result.iloc[:,0]/result.iloc[:,1] * 100
    result = result.rename(index={0:'0 - Q1', 1:'Q1 - Q2', 2:'Q2 - Q3', 3:'Q3 - Q4', 4:'Total'})
    return result
    
    
```
